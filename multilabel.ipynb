```python
import torch
import torch.nn as nn
from sklearn.datasets import make_multilabel_classification
```

```python
X, y = make_multilabel_classification(n_samples=1000, n_features=20, n_classes=5, n_labels=2)
```

```python
class Multilabel(nn.Module):
    def __init__(self, input_size, hidden_size, num_label):
        super(Multilabel, self).__init__()

        self.f1 = nn.Linear(input_size, hidden_size)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.5)


        self.f2 = nn.Linear(hidden_size, hidden_size)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.5)

        self.f3 = nn.Linear(hidden_size, num_label)


    def forward(self, x):
        x = self.dropout1(self.relu1(self.f1(x)))    
        x = self.dropout2(self.relu2(self.f2(x)))    
        x = self.f3(x)    
```

```python
model = Multilabel(5,10,5)
```

```python
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.)
```
